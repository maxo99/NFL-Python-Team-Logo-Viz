```python
# Dependencies
import nfl_data_py as nfl
import pandas as pd
import numpy as np
import urllib.request
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
# Pull the team description 
logos = nfl.import_team_desc()
```

```python
paths = []
```

```python
# Pull the team logos from the URL and save them in the logos folder
for team in range(len(logos)):
    urllib.request.urlretrieve(logos['team_logo_espn'][team], f"logos/{logos['team_abbr'][team]}.tif")
    paths.append(f"logos/{logos['team_abbr'][team]}.tif")
```

```python
plt.rcParams["figure.figsize"] = [20, 14]
plt.rcParams["figure.autolayout"] = True


def getImage(path):
   return OffsetImage(plt.imread(path, format="tif"), zoom=.075)


x = np.random.randint(1, high=100, size=32)
y = np.random.randint(1, high=100, size=32)
fig, ax = plt.subplots()
for x0, y0, path in zip(x, y, paths):
   ab = AnnotationBbox(getImage(path), (x0, y0), frameon=False)
   ax.add_artist(ab)
plt.xticks(range(100));
plt.yticks(range(100));
plt.savefig("hello")
```
